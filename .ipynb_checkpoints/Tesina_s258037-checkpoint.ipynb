<a id=audio_fun></a>
**Audio transformation functions**

```python
def add_noise(data, noise_factor):
    noise = np.random.randn(len(data))
    aug_data = data + noise_factor * noise
    aug_data = aug_data.astype(type(data[0]))
    return aug_data

def stretch(data, speed_factor):
    return librosa.effects.time_stretch(data, speed_factor)

def pitch_change(data, rate, pitch):
    return librosa.effects.pitch_shift(data, rate, pitch)

def save_changes(noise, stretch, pitch, i, sampling_rate):
    librosa.output.write_wav("noisy"+str(i)+".wav", noise, sampling_rate)
    librosa.output.write_wav("stretchy"+str(i)+".wav", stretch, sampling_rate)
    librosa.output.write_wav("pitchy"+str(i)+".wav", pitch, sampling_rate)
```

### 3. Univariate analysis: discover outliers
[[return to top](#Table-of-Contents)]

Detecting and handling outliers is one of the important things to do in classification. Some classification methods are affected by them, so by removing them accuracy can be improved.

To detect them I will use the **1.5 InterQuantile Range rule** (IQR). It allows to take only the results between the first quantile and the third quantile, the central values of the measurements. Everything before (first quantile * 1.5) and after (third quantile * 1.5) is an outlier.

See [analysis functions](#analysis_fun)

#### 6.A. K-Nearest Neighbors
[[return to top](#Table-of-Contents)]

This is the simplest non-parametric classifier. It takes K points in the training set that are nearest to the text input x and counts how many members of each class are in this set. The result is the estimate as an empirical fraction:

<center>$ P(y=c|x, D, K) = \frac{1}{K} \sum_{i \in N_K (x,D)} (y_i = c)$</center>

where $N_K (x,D)$ are the K nearest points to **x** in D.

```python
clf_knn=clf_accuracy(KNeighborsClassifier())
plot_clf(clf_knn, 'KNN')
```

It is possible to change the number of neighbors considered in the algorithm in order to improve accuracy. The search for the best K will be done in [parameter tuning](#7.-Parameter-tuning)

#### 6.B. Logistic Regression
[[return to top](#Table-of-Contents)]

Logistic regression is different from K-NN because it is a parametric classifier, so it is faster and does not explode in case of very large dataset. In this case, due to the number of classes, **multinomial regression** is used. The formula is

<font size="3"><center>$ P(y=c|x) = \frac{e^{\beta_{0k} + \beta_{1k} X_1 + ... + \beta_{pk} X_p}}{\sum_{\ell=1}^C e^{\beta_{0\ell} + \beta_{1\ell} X_1 + ... + \beta_{p\ell} X_p}} $</center></font>

It returns always a value between 0 and 1. The $\beta$ parameters calculated in relation to the **likelihood** of the data. The likelihood is the probability of observing the classes in the data. The parameters that gives the maximum likelihood are chosen.

```python
clf_lr=clf_accuracy(LogisticRegression())
plot_clf(clf_lr, 'Logistic regression')
```

#### 6.C. Support Vector Classifier
[[return to top](#Table-of-Contents)]

The goal of this method is finding an **hyperplane** that separates data according to the label as well as possible. An hyperplane as this form:

<center>$ \beta_0 +\beta_1X_1 + \beta_2X_2 + ... + \beta_pX_p = 0$</center>

where $p$ is the number of dimensions and $(\beta_1,\beta_2,...\beta_p)$ is the **normal vector** of the plane.
The maximum margin between the plane and the data is to be find in order to separate data very well. 

However this is not always possible because of noisy or unseparable data. So the support vector classifier tries to maximize a **soft** margin, a better wider margin at cost of some misclassification error. The formula is

<center>$\underset{\beta_0,\beta_1,...\beta_p, \epsilon_0,...\epsilon_n}{\textrm{maximize}} M \space  \textrm{subject to} \sum_{j=1}^p \beta_j^2 = 1, $</center>

<center>$y_i (\beta_0 + \beta_1x_{i1}+\beta_2x_{i2}+...+\beta_px_{ip}) \ge M(1-\epsilon_i), \epsilon_i\ge0, \sum_{i=1}^n \le C$</center>

where C is a **regularization parameter**, the trade-off between maximizing the margin and minimizing the mistakes. If C is large, SVC focuses on maintaint the misclassification low, while if it is small the margin gains more importance.

```python
clf_svm=clf_accuracy(SVC(kernel='linear'))
```

```python
plot_clf(clf_svm, 'SVM with linear kernel')
```

As expected, a linear kernel performs not as good as logistic regression.  
To understand what a kernel function is, the **inner product between vectors** must be defined.

<center>$\langle x, x_{i'} \rangle = \sum_{j=1}^p x_{ij}x_{i'j} $</center>

and the linear SVC could be rewritten as

<center>$ f(x) = \beta_0 + \sum_{i=1}^n{\alpha_i \langle x, x_i \rangle} $</center>

but computing the inner product can be very abstract, so we can use different kernel functions according to the distribution of data.
For example in this case maybe is more useful a **Radial Kernel** due to the fact that there are many dimensions.
Here is the formula:

<center>$K(x_i,x_{i'}) = exp(-\gamma \sum_{j=1}^p (x_{ij}-x_{i'j})^2) $</center>

And so $ f(x) = \beta_0 + \sum_{i \in S} \hat{\alpha}_i K(x,x_i)$

```python
clf_svm=clf_accuracy(SVC(kernel='rbf'))
```

```python
plot_clf(clf_svm, 'SVM with radial kernel')
```

Much better result this time.
Moreover, due to the fact that there are three classes, SVC works using the **OVO** method. It stands for One versus One and fits all $K\choose 2$ classifers $\hat{f}_{kl}(x)$. Then it classify $x^*$ to the class that wins most pairwise competitions.
K, the number of classes, is not too high (K=3), so it is more convenient OVO

#### 6.D. Decision tree
[[return to top](#Table-of-Contents)]

The Decision Tree classifier applies a recursive binary splitting, stopping when each leaf has fewer than a minimum amount of samples. The criterion that can be used is the **error classifaction rate**, the fraction of observed samples that do not belong to the most common class:

<center>$ E = 1 - \underset{k}{\textrm{max}}(\hat{p}_{mk}) $</center>

where $ \hat{p}_{mk} $ is the proportion of training observations in the $m$-th region that are from the $k$ class. But this is not sufficiently sensible for growing a tree, so another measure is use, the **Gini Index**.

The Gini index is a measure of **purity** of one node of the tree. It is a value between 0 and 1 and when it is very small, the node contains more samples of one class than the others. So when it reaches 0, the tree stops growing from that leaf. The formula is

<center>$ G = \sum_{k=1}^K \hat{p}_{mk}(1-\hat{p}_{mk})$</center>

```python
clf_tree=clf_accuracy(DecisionTreeClassifier(random_state=40))
```

```python
my_plot_tree(clf_tree)
```

The tree that has been created may lead to **overfitting**: it is too precise in relation to the training set. The accuracy can be improved by **pruning** the tree for example. It can be pruned in many ways:
* a **max depth** can be defined

```python
clf_tree=clf_accuracy(DecisionTreeClassifier(random_state=40, max_depth=7))
```

#### 6.D.a Random Forests
[[return to top](#Table-of-Contents)]

This is a better version of the decision tree classifier. It exploit the **boostrap aggregation** or **bagging** procedure. 

This method generates $B$ different bootstrapped training data set from repeated samples from a single training data set. For each test observation on the $b_{th}$ tree, a **majority vote** is taken: the predicted class is the one that occurs the most among the $B$ predictions.

Random forests provide an improvement by  **decorrelating** trees, reducing furthermore the variance. Each time a split in a tree is considered, a random selection of $m$ predictors is chosen as split candidates from the full set of $p$ predictors. Usually $ m \approx \sqrt{p}$

```python
clf_rf=clf_accuracy(RandomForestClassifier(random_state=40))
```

```python
plot_clf(clf_rf, 'Random Forest')
```

### 7. Parameter tuning

For all these classification methods, a best set of parameters can be found using **Grid Search** with **cross validation**. 

A list of parameters for each type of estimator is declared, then the Grid Search tries all the possible combinations of them and returns the best values according to the type of score selected. In this case, the score selected is the accuracy. See [classification functions](#clf_fun)

```python
param_grid_svm={'C':[0.1,1,10,100],'kernel':['linear','rbf']}
grid_search_clf(param_grid_svm, SVC())
```
