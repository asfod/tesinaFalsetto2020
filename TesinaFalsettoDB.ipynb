```r
library(seewave)
library(tuneR)
library(fftw)

normalize <- function(x) {
    return ((x-min(x))/(max(x)-min(x)))
}
filelist <- list.files(pattern = '.*.wav')
label='falsetto'
df <- data.frame(meanfreq=numeric(),sd=numeric(),median=numeric(),Q25=numeric(),Q75=numeric(),IQR=numeric(),skew=numeric(),kurt=numeric(),sp.ent=numeric(),sfm=numeric(),mode=numeric(),centroid=numeric(),meanfun=numeric(),minfun=numeric(),maxfun=numeric(),meandom=numeric(),mindom=numeric(),maxdom=numeric(),dfrange=numeric(),modindx=numeric(),label=character(), stringsAsFactors = FALSE)
#cbind(df,list(meanfreq=numeric(),sd=numeric(),median=numeric(),Q25=numeric(),Q75=numeric(),IQR=numeric(),skew=numeric(),kurt=numeric(),sp.ent=numeric(),sfm=numeric(),mode=numeric(),centroid=numeric(),peakf=numeric(),meanfun=numeric(),minfun=numeric(),maxfun=numeric(),meandom=numeric(),mindom=numeric(),maxdom=numeric(),dfrange=numeric(),label=character()))
for(file in filelist) {
    wavObj <- readWave(file)
    samplespec <- spec(wavObj, f = wavObj@samp.rate, plot = F)
    sp <- specprop(samplespec, f = wavObj@samp.rate, flim = c(0, 280/1000), plot = F)
    
    ff <- fund(wavObj, f = wavObj@samp.rate, ovlp = 50, threshold = 5, fmax = 280, ylim=c(0, 280/1000), plot = FALSE, wl = 2048)[, 2]
    meanfun <- mean(ff, na.rm = TRUE)
    minfun <- min(ff, na.rm = TRUE)
    maxfun <-max(ff, na.rm = TRUE)
    
    y <- dfreq(wavObj, f = wavObj@samp.rate, wl = 2048, ylim=c(0, 280/1000), ovlp = 0, plot = F, threshold = 5, bandpass = c(0,22) * 1000, fftw = TRUE)[,2]
    meandom <- mean(y, na.rm = TRUE)
    mindom <- min(y, na.rm = TRUE)
    maxdom <- max(y, na.rm = TRUE)
    dfrange <- (maxdom - mindom)
    
    #Calculate modindx
    changes <- vector()
    for(j in which(!is.na(y))){
      change <- abs(y[j] - y[j + 1])
      changes <- append(changes, change)
    }
    if(mindom==maxdom) modindx<-0 else modindx <- mean(changes, na.rm = T)/dfrange
    
    

    
    #fp <- fpeaks(a, nmax=1); #dominant frequency: two column matrix, x and y coordinates (frequency in fp[1])
    df <- rbind(df,list(meanfreq=sp$mean/1000,sd=sp$sd/1000,median=sp$median/1000,Q25=sp$Q25/1000,Q75=sp$Q75/1000,IQR=sp$IQR/1000,skew=sp$skewness,kurt=sp$kurtosis,sp.ent=sp$sh,sfm=sp$sfm,mode=sp$mode/1000,centroid=sp$cent/1000,meanfun=meanfun,minfun=minfun,maxfun=maxfun,meandom=meandom,maxdom=maxdom,mindom=mindom,dfrange=dfrange,modindx=modindx,label=label))
}
```
